```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

## Counts of Job Titles

```python
# Counting the number of job postings for each job title
df.pivot_table(index='job_title_short', aggfunc='size')
```

```python
# counting number of job postings for each job title using groupby
df.groupby('job_title_short').size()
```

```python
# calculating the median salary for each job title
df.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc='median')
```

```python
df.pivot_table(values='salary_year_avg', index='job_title_short', columns='job_country',aggfunc = 'median')
# now there are more coulumns than rows which makes the output more difficult toread. Better to have fewer columns. We can do this by setting
# the index to job_title_short and the columns to job_country. This will make the output easier to read.
```

```python
# pivot table showing average yearly salary for each job title and country
df.pivot_table(values='salary_year_avg', index='job_country', columns='job_title_short',aggfunc = 'median')
# output is easier to read. 
```

```python
# top 6 countries with the most job postings arranged in an array, setting variable to top_countries
top_countries = df['job_country'].value_counts().head(6).index
top_countries
```

```python
top_countries = df['job_country'].value_counts().head(6).index

# create a pivot table showing the median salary for each job title and country, but only for the top 6 countries with the most job postings. Set the index to job_country and the columns to job_title_short. 
# Set the variable to df_job_country_salary.

df_job_country_salary = df.pivot_table(
    values='salary_year_avg', 
    index='job_country', 
    columns='job_title_short',
    aggfunc='median'
)

# filter the dataframe on top_countries by setting the df equal to itself and filtering 
# using the .loc method

df_job_country_salary = df_job_country_salary.loc[top_countries]

job_titles = ['Data Analyst', 'Data Engineer', 'Data Scientist']

# then we filter on specific job_titles by passing in the list of job_titles 
                                                  
df_job_country_salary = df_job_country_salary[job_titles]

df_job_country_salary
```

```python
df_job_country_salary.plot(kind='bar', figsize=(10,6))
plt.title('Median Salary by Job Title and Country')
```
