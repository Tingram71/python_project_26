```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_us = df[df['job_country'] == 'United States'].copy()
```

```python
df_us['job_posted_month'] = df_us['job_posted_date'].dt.strftime('%B')

df_us
```

```python
# now we need to pivot the data in order to get it in the right format for plotting.
# month along the index. columns wll be job titles.

df_us_pivot = df_us.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_us_pivot
```

```python
# we need to sort the months chronologically. We have to look at the month number and sort in numerical order.

df_us_pivot.reset_index(inplace=True)
df_us_pivot['month_no'] = pd.to_datetime(df_us_pivot['job_posted_month'], format = '%B').dt.month

```

```python
df_us_pivot
```

```python
df_us_pivot.sort_values('month_no', inplace=True)
df_us_pivot.set_index('job_posted_month', inplace=True)
```

```python
df_us_pivot.drop(columns='month_no', inplace=True)
```

```python
df_us_pivot
```

```python
df_us_pivot.plot(kind='line')
plt.title('Job demand in the US by month')
plt.xlabel('Month')
```

```python
# lets clean up the graph to show the top three jobs
top_3 = df_us['job_title_short'].value_counts().head(3)
top_3

```

```python
top_3 = top_3.index.to_list()
top_3 
```

```python
df_us_pivot[top_3].plot(kind='line'), df_us_pivot[top_3].plot(kind='bar')
```
